```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "/Users/Admin/Documents/DataBootCampRepo/GitLab/UCBER201805DATA2-Class-Repository-DATA/02-Homework/05-Matplotlib/Instructions/Pyber/data/city_data.csv"
ride_data_to_load = "/Users/Admin/Documents/DataBootCampRepo/GitLab/UCBER201805DATA2-Class-Repository-DATA/02-Homework/05-Matplotlib/Instructions/Pyber/data/ride_data.csv"

# Read the City and Ride Data

```

```python
# Use Pandas to read data in CSV file
city_data_to_load = pd.read_csv(city_data_to_load)

# Print the first few results from the file
city_data_to_load.head()
```

```python
ride_data_to_load = pd.read_csv(ride_data_to_load)
ride_data_to_load.head()
```

```python
# Combine the data into a single dataset

# Display the data table for preview

```

```python
# Merge two dataframes using an inner join
merge_table = pd.merge(city_data_to_load, ride_data_to_load, on="city")
merge_table.head(10)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
#Find the average fare per city.

average_fare = merge_table.groupby("city")["fare"].mean()
average_fare.head()
```

```python
#Find the total rides per city.

total_rides = merge_table.groupby("city")["ride_id"].count()
total_rides.head()
```

```python
#Find city type
city_type = merge_table.set_index('city')['type']

city_type.value_counts()

```

```python
#Find the total number of drivers per city.

total_drivers = merge_table.groupby("city")["driver_count"].count()
total_drivers
```

```python
# data frames for city types - Urban, Suburban and Rural
urban_city = merge_table.loc[(merge_table["type"] == "Urban")]
suburban_city = merge_table.loc[(merge_table["type"] == "Suburban")]
rural_city = merge_table.loc[(merge_table["type"] == "Rural")]

# Urban: Per city Average Fare, Total Rides and Total Drivers 
avg_fare_urban_city = urban_city.groupby(['city'])['fare'].mean()
total_rides_urban_city = urban_city.groupby(['city'])['ride_id'].count()
total_drivers_urban_city = urban_city.groupby(['city'])['driver_count'].count()

# Suburban: Per city Average Fare, Total Rides and Total Drivers 
avg_fare_suburban_city = suburban_city.groupby(['city'])['fare'].mean()
total_rides_suburban_city = suburban_city.groupby(['city'])['ride_id'].count()
#total_rides_suburban_city = suburban_city.groupby(['city']).count()['ride_id']
total_drivers_suburban_city = suburban_city.groupby(['city'])['driver_count'].count()

# Rural: Average Fare, Total Rides and Total Drivers for Rural cities
avg_fare_rural_city = rural_city.groupby(['city'])['fare'].mean()
total_rides_rural_city = rural_city.groupby(['city'])['ride_id'].count()
total_drivers_rural_city = rural_city.groupby(['city'])['driver_count'].count()
```

```python
# Create scatter plots for Urban, Suburban and Rural cities
plt.scatter(total_rides_urban_city, avg_fare_urban_city,s=total_drivers_urban_city*15,
            marker ='o', facecolors ="coral", edgecolors='black',alpha = 0.5, label="Urban")

plt.scatter(total_rides_suburban_city, avg_fare_suburban_city,s=total_drivers_suburban_city*15,
            marker ='o', facecolors ="lightskyblue", edgecolors='black',alpha = 0.5, label="Suburban")

plt.scatter(total_rides_rural_city, avg_fare_rural_city,s=total_drivers_rural_city*15,
            marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.5, label="Rural")

#Add title, x axis label and y axis label.
plt.title("Average Fare per City versus Number of Rides")
plt.xlabel("Number of Rides")
plt.ylabel("Average Fare Per City")
plt.text(40, 50,"Note: Circle size correlates with driver count per city.")

#Add the legend.
plt.legend(loc= "upper right")

#Add gridlines
plt.grid(color='gray', linestyle='-', linewidth=.5)
```
